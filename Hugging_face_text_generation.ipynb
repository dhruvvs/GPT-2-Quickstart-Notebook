# **SIMPLE TEXT GENERATION USING PIPELINE API**

```python
from transformers import pipeline

# Load GPT-2 using pipeline
generator = pipeline("text-generation", model="gpt2")

prompt = "In a future where AI helps everyone,"
outputs = generator(prompt, max_new_tokens=100, do_sample=True, temperature=0.9, top_p=0.95)

# Print outputs
for i, o in enumerate(outputs):
    print(f"--- Result {i+1} ---")
    print(o["generated_text"])
```

# **ADVANCED TEXT GENERATION USING CUSTOM MODELINING AND CONFIG
bold text

```python
from transformers import AutoTokenizer, AutoModelForCausalLM, GenerationConfig

model_name = "gpt2"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)

prompt = "In a future where AI helps everyone,"
inputs = tokenizer(prompt, return_tensors="pt")

gen_config = GenerationConfig(
    max_new_tokens=100,
    temperature=0.9,
    top_p=0.95,
    do_sample=True,
    num_return_sequences=2
)

outputs = model.generate(**inputs, generation_config=gen_config)

# Decode and print
for i, output in enumerate(outputs):
    print(f"--- Advanced Result {i+1} ---")
    print(tokenizer.decode(output, skip_special_tokens=True))

```
